```python
import folium
```

```python
import requests
from pandas import json_normalize
```

```python
loc = [37.5662952, 126.9779451] #위도 경도
```

```python
map = folium.Map(location=loc, zoom_start=16)
folium.Marker(location=loc).add_to(map)
map
```

```python
targetSite='https://www.starbucks.co.kr/store/getStore.do?r=8K3XF06R9Q'
req=requests.post(targetSite, data={
'ins_lat': 37.563398,
'ins_lng': 126.9863309,
'p_sido_cd': '01',
'p_gugun_cd': '',
'in_biz_cd': '',
'iend': 600,
'set_date': ''
})
starbucks=req.json()
print(type(starbucks))
```

```python
st_df = json_normalize(starbucks, 'list')
st_df
```

```python
st_df_map = st_df[['s_name','sido_code','sido_name','gugun_code','gugun_name',
'doro_address','lat','lot']]
st_df_map
```

```python
st_df_map['lat'] = st_df_map['lat'].astype(float)
st_df_map['lot'] = st_df_map['lot'].astype(float)
```

```python
st_df_map.info()
```

```python
loc = [37.501087, 127.043069]
stMap = folium.Map(location=loc, zoom_start=18)
pop = folium.Popup('역삼아레나빌딩', max_width=100)
folium.Marker(location=loc, popup=pop).add_to(stMap)
stMap
```

```python
loc_a = []
for i in range(0,100):
    a = st_df_map['lat'][i]
    b = st_df_map['lot'][i]
    loc_a.append([a,b])
    
stMap = folium.Map(location=loc, zoom_start=18)
for cor in loc_a:
    pop = folium.Popup('star', max_width=100)
    folium.Marker(location=cor, popup=pop).add_to(stMap)
stMap
```
